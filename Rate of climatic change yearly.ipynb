```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,4)) # increase the size 18 for row,4 for the columns

import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate2.plot(kind="bar", x="Year",y=["Doubling Time (Dt)"])
```

```python
x=climate2[['Year','Total no. of publication with %','Cumulative','Natural log','Relative Growth Rate']]
```

```python
y=climate2['Doubling Time (Dt)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression().fit(x_train,y_train)
```

```python
model.predict(x_test).round()
```

```python
y_test
```

```python
prediction=model.predict(x_test).round()
```

```python
from sklearn.metrics import r2_score
r2_score(prediction, y_test)*100
```
